```python
import os
# ...existing code...
def split_players_data(input_file='../data/cuiaba.csv', output_directory='player_files'):
    """
    Split the cuiaba.csv file into separate files for each player.
    
    Args:
        input_file (str): Path to the input CSV file
        output_directory (str): Directory where individual player files will be saved
    """
    
    # Create output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
    
    # Read the CSV file
    df = pd.read_csv(input_file)
    
    # Strip leading/trailing whitespace from column names
    df.columns = df.columns.str.strip()
    
    # Get unique players
    unique_players = df['Player Name'].unique()
    
    print(f"Found {len(unique_players)} unique players")
    print(f"Total rows in original file: {len(df)}")
    print(f"Cleaned column names: {df.columns[:10].tolist()}")  # Show first 10 cleaned columns
    
    # Split data for each player
    for player in unique_players:
        # Filter data for current player
        player_data = df[df['Player Name'] == player]
        
        # Create filename (replace any problematic characters)
        safe_player_name = str(player).replace('/', '_').replace('\\', '_').replace(':', '_')
        filename = f"player_{safe_player_name}.csv"
        filepath = os.path.join(output_directory, filename)
        
        # Save player data to separate file
        player_data.to_csv(filepath, index=False)
        
        print(f"Created {filename} with {len(player_data)} rows")
    
    print(f"\nAll files saved in '{output_directory}' directory")

# Run the function
split_players_data()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime

# Set style for better plots
plt.style.use('default')
sns.set_palette("husl")

# Load and examine the data
df = pd.read_csv('player_files/player_1.csv')

print("Dataset Overview:")
print(f"Shape: {df.shape}")
print(f"\nColumns: {len(df.columns)} total columns")
print(f"\nFirst few column names:")
print(df.columns[:20].tolist())

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Display basic info about the matches
print(f"\nMatches data from {df['Date'].min()} to {df['Date'].max()}")
print(f"Total matches: {len(df)}")
print(f"\nMatch types:")
print(df['Period Name'].value_counts())
```

```python
# Key performance metrics analysis
print("Key Performance Metrics Summary:")
print("="*50)

# Select key metrics for analysis
key_metrics = [
    'Odometer',  # Distance covered
    'Player Load',  # Physical load
    'Max Velocity',  # Top speed
    'Avg Heart Rate',  # Average heart rate
    'Max Heart Rate',  # Maximum heart rate
    'Field Time%',  # Time on field percentage
    'Duration'  # Match duration
]

# Clean and convert duration to minutes
def duration_to_minutes(duration_str):
    try:
        if pd.isna(duration_str):
            return 0
        # Handle format like '02:36:26' or '00:48:57'
        parts = str(duration_str).split(':')
        if len(parts) == 3:
            return int(parts[0]) * 60 + int(parts[1]) + int(parts[2])/60
        return 0
    except:
        return 0

df['Duration_minutes'] = df['Duration'].apply(duration_to_minutes)

# Convert string columns to numeric
numeric_cols = ['Odometer', 'Player Load', 'Max Velocity', 'Avg Heart Rate', 'Max Heart Rate']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert Field Time% (remove % and convert to float)
df['Field Time%'] = df['Field Time%'].astype(str).str.replace(',', '.').astype(float)

# Summary statistics
summary_stats = df[numeric_cols + ['Field Time%', 'Duration_minutes']].describe()
print(summary_stats)
```

```python
# Create comprehensive visualizations
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Player Performance Analysis Over Time', fontsize=16, fontweight='bold')

# 1. Distance covered over time
axes[0,0].plot(df['Date'], df['Odometer'], marker='o', linewidth=2, markersize=6)
axes[0,0].set_title('Distance Covered per Match (meters)', fontweight='bold')
axes[0,0].set_ylabel('Distance (m)')
axes[0,0].tick_params(axis='x', rotation=45)
axes[0,0].grid(True, alpha=0.3)

# 2. Player Load over time
axes[0,1].plot(df['Date'], df['Player Load'], marker='s', color='orange', linewidth=2, markersize=6)
axes[0,1].set_title('Player Load per Match', fontweight='bold')
axes[0,1].set_ylabel('Player Load')
axes[0,1].tick_params(axis='x', rotation=45)
axes[0,1].grid(True, alpha=0.3)

# 3. Max Velocity over time
axes[0,2].plot(df['Date'], df['Max Velocity'], marker='^', color='red', linewidth=2, markersize=6)
axes[0,2].set_title('Maximum Velocity per Match (km/h)', fontweight='bold')
axes[0,2].set_ylabel('Max Velocity (km/h)')
axes[0,2].tick_params(axis='x', rotation=45)
axes[0,2].grid(True, alpha=0.3)

# 4. Heart Rate comparison
axes[1,0].plot(df['Date'], df['Avg Heart Rate'], marker='o', label='Average HR', linewidth=2)
axes[1,0].plot(df['Date'], df['Max Heart Rate'], marker='s', label='Max HR', linewidth=2)
axes[1,0].set_title('Heart Rate Analysis', fontweight='bold')
axes[1,0].set_ylabel('Heart Rate (bpm)')
axes[1,0].legend()
axes[1,0].tick_params(axis='x', rotation=45)
axes[1,0].grid(True, alpha=0.3)

# 5. Field Time Percentage
axes[1,1].bar(range(len(df)), df['Field Time%'], color='green', alpha=0.7)
axes[1,1].set_title('Field Time Percentage per Match', fontweight='bold')
axes[1,1].set_ylabel('Field Time %')
axes[1,1].set_xlabel('Match Number')
axes[1,1].grid(True, alpha=0.3)

# 6. Match Duration
axes[1,2].bar(range(len(df)), df['Duration_minutes'], color='purple', alpha=0.7)
axes[1,2].set_title('Match Duration (minutes)', fontweight='bold')
axes[1,2].set_ylabel('Duration (minutes)')
axes[1,2].set_xlabel('Match Number')
axes[1,2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Correlation analysis between key metrics
correlation_metrics = ['Odometer', 'Player Load', 'Max Velocity', 'Avg Heart Rate', 
                      'Max Heart Rate', 'Field Time%', 'Duration_minutes']

# Create correlation matrix
corr_matrix = df[correlation_metrics].corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Correlation Matrix - Key Performance Metrics', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nStrong Correlations (>0.7 or <-0.7):")
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        corr_val = corr_matrix.iloc[i, j]
        if abs(corr_val) > 0.7:
            print(f"{corr_matrix.columns[i]} vs {corr_matrix.columns[j]}: {corr_val:.3f}")
```

```python
# Generate detailed insights
print("DETAILED PERFORMANCE INSIGHTS")
print("="*50)

# Overall performance summary
total_distance = df['Odometer'].sum()
avg_distance_per_match = df['Odometer'].mean()
max_distance = df['Odometer'].max()
total_playtime = df['Duration_minutes'].sum()

print(f"\n📊 OVERALL STATISTICS:")
print(f"• Total distance covered: {total_distance:,.0f} meters ({total_distance/1000:.1f} km)")
print(f"• Average distance per match: {avg_distance_per_match:,.0f} meters")
print(f"• Best performance (distance): {max_distance:,.0f} meters")
print(f"• Total playing time: {total_playtime:.0f} minutes ({total_playtime/60:.1f} hours)")

# Performance by match type
print(f"\n⚽ PERFORMANCE BY MATCH TYPE:")
for period in df['Period Name'].unique():
    if pd.notna(period):
        period_data = df[df['Period Name'] == period]
        avg_dist = period_data['Odometer'].mean()
        avg_load = period_data['Player Load'].mean()
        print(f"• {period}: Avg distance {avg_dist:,.0f}m, Avg load {avg_load:.1f}")

# Best and worst performances
best_distance_idx = df['Odometer'].idxmax()
worst_distance_idx = df['Odometer'].idxmin()

print(f"\n🏆 BEST PERFORMANCE:")
print(f"• Date: {df.loc[best_distance_idx, 'Date'].strftime('%Y-%m-%d')}")
print(f"• Distance: {df.loc[best_distance_idx, 'Odometer']:,.0f} meters")
print(f"• Player Load: {df.loc[best_distance_idx, 'Player Load']:.1f}")
print(f"• Max Velocity: {df.loc[best_distance_idx, 'Max Velocity']:.1f} km/h")

print(f"\n📉 LOWEST PERFORMANCE:")
print(f"• Date: {df.loc[worst_distance_idx, 'Date'].strftime('%Y-%m-%d')}")
print(f"• Distance: {df.loc[worst_distance_idx, 'Odometer']:,.0f} meters")
print(f"• Player Load: {df.loc[worst_distance_idx, 'Player Load']:.1f}")

# Fitness trends
if len(df) >= 10:
    recent_matches = df.tail(5)
    early_matches = df.head(5)
    
    recent_avg_dist = recent_matches['Odometer'].mean()
    early_avg_dist = early_matches['Odometer'].mean()
    trend = ((recent_avg_dist - early_avg_dist) / early_avg_dist) * 100
    
    print(f"\n📈 FITNESS TREND:")
    print(f"• Early season avg distance: {early_avg_dist:,.0f}m")
    print(f"• Recent matches avg distance: {recent_avg_dist:,.0f}m")
    print(f"• Trend: {trend:+.1f}% {'improvement' if trend > 0 else 'decline'}")
```
