```python
import pandas as pd
import os

def split_players_data(input_file='../data/cuiaba.csv', output_directory='player_files'):
    """
    Split the cuiaba.csv file into separate files for each player.
    
    Args:
        input_file (str): Path to the input CSV file
        output_directory (str): Directory where individual player files will be saved
    """
    
    # Create output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
    
    # Read the CSV file
    df = pd.read_csv(input_file)
    
    # Get unique players
    unique_players = df['Player Name'].unique()
    
    print(f"Found {len(unique_players)} unique players")
    print(f"Total rows in original file: {len(df)}")
    
    # Split data for each player
    for player in unique_players:
        # Filter data for current player
        player_data = df[df['Player Name'] == player]
        
        # Create filename (replace any problematic characters)
        safe_player_name = str(player).replace('/', '_').replace('\\', '_').replace(':', '_')
        filename = f"player_{safe_player_name}.csv"
        filepath = os.path.join(output_directory, filename)
        
        # Save player data to separate file
        player_data.to_csv(filepath, index=False)
        
        print(f"Created {filename} with {len(player_data)} rows")
    
    print(f"\nAll files saved in '{output_directory}' directory")

# Run the function
split_players_data()
```
